```python
import pandas as pd

# Carga el dataset en un DataFrame de pandas
df = pd.read_csv('titanic.csv')

# Esto te permite ver si los nombres de las columnas están bien
df.head()
# Utiliza el comando df.describe() para obtener estadísticas de los datos
df.describe()
```

```python
# 4.1 Análisis de supervivencia
# Calculamos el promedio de la columna 'Survived'
sobrevivientes = df['Survived'].mean() * 100
print(f"1. Porcentaje de supervivencia de los pasajeros: {sobrevivientes:.2f}%")

# 4.2 Análisis de clase (Parch)
print("\n2. Porcentaje de pasajeros según cada clase (Parch):")
print(df['Parch'].value_counts(normalize=True) * 100)

# 4.3 Análisis por Género
print("\n3. Cantidad de hombres y mujeres que viajaron:")
print(df['Sex'].value_counts())
```
